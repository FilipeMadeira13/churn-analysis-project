# Limpeza de dados

### Correções para realizar nesse dataset:

##### 1. Coluna TotalCharges como numérico

```python
import pandas as pd

df = pd.read_csv('../data/raw/churn.csv')

df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
```

##### 2. . Converter churn para numérico

```python
df['Churn'] = df['Churn'].map({'Yes': 1, 'No': 0})
```

##### Salvando dados limpos

```python
df.to_csv("../data/processed/churn_clean.csv", index=False)
```
